# Imports

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

# YOLO11 (100eph, 0.01lr)
## Train - HBCD 

```python
f1_path = os.path.normpath("../runs/test/wdf_test/fold_1/hbcd/yolo11n_100eph_001lr/results.csv")
f2_path = os.path.normpath("../runs/test/wdf_test/fold_2/hbcd/yolo11n_100eph_001lr/results.csv")

# 1. Load your results
fold1 = pd.read_csv(f1_path)
fold2 = pd.read_csv(f2_path)

# Clean column names (remove leading/trailing spaces)
fold1.columns = [c.strip() for c in fold1.columns]
fold2.columns = [c.strip() for c in fold2.columns]

# 2. Calculate the Average (Mean) of both dataframes
cv_avg = (fold1 + fold2) / 2

# 3. Define the metrics we want to plot (matching YOLO's results.png)
# Note: Check your CSV column names; older versions might use 'metrics/mAP_0.5'
metrics_to_plot = [
    'metrics/precision(B)', 'metrics/recall(B)',
    'metrics/mAP50(B)', 'metrics/mAP50-95(B)'
]

# 4. Create the Figure (2 rows, 5 columns to fit all 10 common metrics)
fig, axes = plt.subplots(2, 2)
axes = axes.ravel()

for i, metric in enumerate(metrics_to_plot):
    if metric in cv_avg.columns:
        axes[i].plot(cv_avg['epoch'], cv_avg[metric], color='blue', linewidth=2)
        axes[i].set_title(metric.replace('/', '\n'), fontsize=10)
        axes[i].grid(True, linestyle='--', alpha=0.6)
    else:
        axes[i].set_visible(False) # Hide if metric name doesn't match exactly

plt.tight_layout()
plt.show()

print(f"Final CV Precision: {cv_avg['metrics/precision(B)'].iloc[-1]:.4f}")
print(f"Final CV Recall: {cv_avg['metrics/recall(B)'].iloc[-1]:.4f}")
print(f"Final CV mAP50: {cv_avg['metrics/mAP50(B)'].iloc[-1]:.4f}")
print(f"Final CV mAP50-95: {cv_avg['metrics/mAP50-95(B)'].iloc[-1]:.4f}")
```
