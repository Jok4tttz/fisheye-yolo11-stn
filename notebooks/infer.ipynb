# Imports

```python
import os
import torch
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from ultralytics import YOLO
```

# Train Image

```python
model = YOLO(os.path.normpath("../models/trained/HBCD/yolo11n_stn_fold1_1keph_01lr.pt"))
source = os.path.normpath("../datasets/images/wdf_test/fold_1/train/call_center_000001.jpg")
ROI = os.path.normpath("../datasources/WEPDTOF/ROI_maps/tech_store.png")
with Image.open(source) as img:
        # Get width and height
        width, height = img.size
        
        print(f"Image Dimensions: {width} x {height} pixels")
        print(f"Image Width: {width}px")
        print(f"Image Height: {height}px\n")

with Image.open(ROI) as img:
        # Get width and height
        width, height = img.size
        
        print(f"ROI Dimensions: {width} x {height} pixels")
        print(f"ROI Width: {width}px")
        print(f"ROI Height: {height}px\n")
results = model(source, show = True)
```

```python
label = os.path.normpath("../datasets/labels/wdf_test/fold_1/train/call_center_000001.txt")

# 2. Get the first result (since we only passed one image)
res = results[0]
with open(label, 'r') as file:
    for count, line in enumerate(file, 1):
        pass

print(f"Infer count: {len(res)}")
print(f"Label count: {count}")

# 3. Plot the result and convert BGR to RGB for Matplotlib
res_plotted = res.plot()
res_rgb = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)

# 4. Display in the notebook
plt.figure(figsize=(12, 8))
plt.imshow(res_rgb)
plt.axis('off')
plt.show()
```
